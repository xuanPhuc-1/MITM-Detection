```python
from sklearn.datasets import load_digits
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix, classification_report, precision_score
from sklearn import metrics
from sklearn.model_selection import train_test_split

```

```python
#I want to combine the CNN and the MLP model to detect the ARP MitM attack
#i will use 2 1D CNN layers to extract features from the dataset, 2 Max Pooling layers to reduce the size of the output from the CNN layers, 1 Flatten layer to flatten the output from the Max Pooling layers, 1 Dense layer to connect the Flatten layer to the MLP model, 1 Dropout layer to prevent overfitting, 1 Dense layer to connect the Dropout layer to the output layer, 1 output layer with 2 neurons to classify the dataset into 2 classes: normal and ARP MitM attack, 1 Dropout layer to prevent overfitting, 1 Dense layer to connect the Dropout layer to the output layer, 1 output layer with 2 neurons to classify the dataset into 2 classes: normal and ARP MitM attack and 1 Dense layer by using ReLU as the activation

from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv1D, MaxPooling1D
from keras.utils import np_utils

x_train, x_test, y_train, y_test = train_test_split( df, dataset_label, random_state=4)

#build the model
model = Sequential()

model.add(Conv1D(filters=64, kernel_size=5, activation='relu', input_shape=(115,1)))
model.add(MaxPooling1D(pool_size=10))
model.add(Conv1D(filters=64, kernel_size=5, activation='relu'))
model.add(MaxPooling1D(pool_size=2))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))


model.add(Flatten())
model.add(Dense(2, activation='sigmoid'))


#add the model with MLP part, i will use 2 hidden layers, 1 Dropout layer to prevent overfitting, 1 Dense layer by using ReLU as the activation. Then we use Fully Connected Later part to connect the CNN and the MLP model

#compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

#train the model
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=5, batch_size=512)

#evaluate the model
scores = model.evaluate(x_test, y_test, verbose=0)



```

```python
#print the accuracy of the model
print("Accuracy: %.2f%%" % (scores[1]*100))
```

```python
#use confusion matrix to evaluate the model
#use matplotlib to plot the confusion matrix
from sklearn.metrics import confusion_matrix
import itertools

y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred,axis = 1)
y_true = np.argmax(y_test,axis = 1)

confusion_mtx = confusion_matrix(y_true, y_pred_classes)

#plot the confusion matrix
plt.imshow(confusion_mtx, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion matrix')
plt.colorbar()
tick_marks = np.arange(2)
plt.xticks(tick_marks, range(2), rotation=45)
plt.yticks(tick_marks, range(2))
thresh = confusion_mtx.max() / 2.
for i, j in itertools.product(range(confusion_mtx.shape[0]), range(confusion_mtx.shape[1])):
    plt.text(j, i, confusion_mtx[i, j],
             horizontalalignment="center",
             color="white" if confusion_mtx[i, j] > thresh else "black")
plt.tight_layout()
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()

```
